# Week 1: Basic Syntax & Data Types

## 1. The Concept (Simple English)
**What is it?**
Think of variables as labeled boxes where you store information. The "Data Type" is simply the *kind* of information you put in that box. The 4 primitive types in Python are:
* **Strings (`str`):** Text characters wrapped in quotes.
* **Integers (`int`):** Whole numbers.
* **Floats (`float`):** Decimal numbers.
* **Booleans (`bool`):** True or False values.

**Why do we use it in real projects?**
Every application runs on data. If you're building a system like Netflix, you need a String for the movie title, an Integer for the release year, a Float for the star rating, and a Boolean to check if the user is logged in.

## 2. The Syntax & Example
**Syntax Rule:** `variable_name = value`
*(Notice there is no need to declare the data type explicitly. Python figures it out automatically!)*

```python
user_name = "Alice"
user_age = 28
account_balance = 250.50
is_premium = True

print("--- Example Output ---")
print(user_name)
print(account_balance)
```

**Line-by-line explanation:**
* **Line 1:** We create a variable named `user_name` and put the text `"Alice"` inside it.
* **Line 2:** We create `user_age` and store the whole number `28`.
* **Line 3:** We create `account_balance` and store the decimal `250.50`.
* **Line 4:** We create `is_premium` and store the boolean value `True` (Note the capital 'T').
* **Lines 6-8:** We use the `print()` function to display the values in the console.

## 3. Common Pitfalls (CRITICAL)
**What mistakes do beginners usually make with this?**
1.  **Naming errors:** Starting variable names with numbers or using spaces instead of underscores.
2.  **Data type mixing:** Forgetting that `"10"` (with quotes) is a String, not an Integer, which causes errors if you try to do math with it.
3.  **Capitalization:** Using lowercase for boolean values (`true` instead of `True`).

**Bad vs. Good Code Comparison:**

```python
# BAD CODE (These will cause errors):
# 1st_name = "Bob"       # SyntaxError: cannot start with a number
# user age = 25          # SyntaxError: spaces are not allowed in variable names
# is_logged_in = true    # NameError: 'true' is not defined (must be 'True')
# total = "10" + 5       # TypeError: cannot add a string and an integer

# GOOD CODE:
first_name = "Bob"     # Uses an underscore instead of starting with a number
user_age = 25          # Uses snake_case (underscores) for spaces
is_logged_in = True    # Correctly uses a capital 'T'
total = 10 + 5         # Adding two integers properly
```
