# Week 2: Control Flow & Logic

## 1. The Concept (Simple English)
**What is it?**
By default, Python reads your code exactly like a book: from top to bottom, line by line. **Control Flow** allows you to change that order. It lets your program skip certain lines, branch off into different directions, or repeat lines multiple times based on conditions you set.



**Why do we use it in real projects?**
Without control flow, software would be completely static and useless. A real-world application must react to different situations.
* *Example:* **IF** a user enters the right password, log them in. **ELSE IF** they get it wrong but have attempts left, show a warning. **ELSE**, lock the account.

## 2. The Syntax & Example
**The Syntax Rule:**
Python uses the keywords `if`, `elif` (else if), and `else`.
*Crucial Rule:* Python uses **indentation** (spaces or tabs) to know which lines of code belong inside the `if` statement.

```python
user_age = 20
has_ticket = True

# Checking multiple conditions using 'and', '>=', and '=='
if user_age >= 18 and has_ticket == True:
    print("Access Granted: Welcome to the VIP lounge.")
    
elif user_age < 18 and has_ticket == True:
    print("Access Denied: You have a ticket, but you are not old enough.")
    
else:
    print("Access Denied: You do not have a ticket.")
    
print("Have a great day!")
```

**Line-by-Line Explanation:**
* **Lines 1-2:** We set up our starting variables (an Integer and a Boolean).
* **Line 5:** The `if` statement checks two things: Is the age 18 or older (`>=`) **AND** is `has_ticket` equal to `True` (`==`)? Because both are true, it runs the indented code directly below it.
* **Line 6:** This line is indented, so it only runs if Line 5 is true. It prints our welcome message.
* **Lines 8-9:** The `elif` (Else If) provides a secondary check. It only gets checked if the first `if` was false.
* **Lines 11-12:** The `else` acts as the final catch-all. If *none* of the above conditions were true, it runs this block.
* **Line 14:** This line is **not** indented. That means it is outside the control flow block and will run no matter what happens above it.

## 3. Common Pitfalls (CRITICAL)
**What mistakes do beginners usually make with this?**
1. **The Single vs. Double Equals:** Beginners often use `=` (assignment) when they mean `==` (comparison). `x = 5` means "Make x equal to 5". `x == 5` asks the question, "Does x equal 5?"
2. **Indentation Errors:** Forgetting to indent the code inside the `if` block, or indenting inconsistently (mixing spaces and tabs). Python will crash if the indentation isn't perfect.
3. **Missing the Colon:** Forgetting to put a colon (`:`) at the very end of the `if`, `elif`, or `else` line.

**Bad vs. Good Code Comparison:**

```python
# BAD CODE:
# if user_age = 18          # SyntaxError: Used '=' instead of '==' and forgot the ':'
# print("You are an adult") # IndentationError: This line needs to be indented!

# GOOD CODE:
if user_age == 18:          # Uses '==' for comparison and ends with ':'
    print("You are an adult") # Properly indented by 4 spaces
```
